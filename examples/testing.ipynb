```python
import backtester as bt
import numpy as np
stock = bt.Asset(name = 'hi', px = 100, mu = 0, sigma = 0.04)
position = bt.Long([stock],[1.0])
p = bt.Portfolio([position], [1.0])
s = bt.MontecarloSimulation(100,1,p)
r = s.run()
```

```python
print(r)
```

```python
import matplotlib.pyplot as plt
y = r['Returns'][0]
plt.plot(y)
plt.show()
```
